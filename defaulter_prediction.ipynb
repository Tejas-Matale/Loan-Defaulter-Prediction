```python
total_acc_avg = df.groupby('total_acc')['mort_acc'].mean()
```

```python
total_acc_avg
```

```python
def fill_mor(total_acc,mort_acc):
    
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
df['mort_acc']=df.apply(lambda x: fill_mor(x['total_acc'], x['mort_acc']), axis=1)
```

<h2>Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
df = df.drop('loan_status',axis=1)
```

```python
x=df.drop('loan_repaid',axis=1).values
```

```python
y=df['loan_repaid'].values
```

```python
x_train,x_test,y_train,y_test =train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

<h2>Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model = Sequential()

model.add(Dense(78,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(39,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(19,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')
```

```python
model.fit(x=x_train,y=y_train,epochs=25,batch_size=256,
          validation_data=(x_test,y_test))
```

```python
from tensorflow.keras.models import load_model
```

```python
model.save('loan_defaulter.h5')
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
prediction=model.predict(x_test)
```

```python
prediction = model.predict(x_test)
predicted_labels = (prediction > 0.5).astype(int).flatten()
y_test_flat = y_test.flatten()  
print(classification_report(y_test_flat, predicted_labels))
```

```python
predicted_labels = (prediction > 0.5).astype(int).flatten()
y_test_flat = y_test.flatten()  
print(confusion_matrix(y_test_flat, predicted_labels))
```
